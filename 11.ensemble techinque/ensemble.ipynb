# Step 3 - Seperate X and Y(loan_status).

```python
X = df.drop(columns=["id", "loan_status"])
Y = df[["loan_status"]]
```

```python
Y.head()
```

```python
Y.value_counts()
```

```python
Y.value_counts().plot(kind="bar")
```

# Step 4 - Apply Prerpocessing on X.

```python
X.dtypes
```

```python
cat = list(X.columns[X.dtypes == "object"])
cat
```

```python
con = list(X.columns[X.dtypes != "object"])
con
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = make_pipeline(SimpleImputer(strategy="median"), StandardScaler())
```

```python
cat_pipe = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore", sparse_output=False, drop="first"),
)
```

```python
pre = ColumnTransformer([("num", num_pipe, con), ("cat", cat_pipe, cat)]).set_output(
    transform="pandas"
)
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

```python
pre
```

# Step 5 - Apply train test split.

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(
    X_pre, Y, test_size=0.2, random_state=42
)
```

# Step 6 - Model Building.

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

model1 = RandomForestClassifier(
    n_estimators=1,
    max_depth=1,
    min_samples_split=5,
    min_samples_leaf=5,
    criterion="gini",
    random_state=42,
)

model1.fit(xtrain, ytrain)
```

```python
model1.score(xtrain, ytrain)
```

```python
model1.score(xtest, ytest)
```

```python
from sklearn.metrics import f1_score

ypred_test = model1.predict(xtest)
f1_score(ytest, ypred_test, average="macro")
```

# Hyperparameter tuning.

```python
parmas1 = {
    "n_estimators": [50, 100, 200],
    "max_depth": [3, 4, 5],
    "criterion": ["gini", "entropy"],
}
```

```python
from sklearn.model_selection import RandomizedSearchCV

rfc = RandomForestClassifier(random_state=42)
rscv_rfc = RandomizedSearchCV(
    rfc, param_distributions=parmas1, cv=5, scoring="f1_macro", n_jobs=-1
)
rscv_rfc.fit(xtrain, ytrain)
```

```python
best_rfc.score(xtest, ytest)
```

```python
ypred_test1 = best_rfc.predict(xtest)
ypred_test1[0:5]
```

```python
from sklearn.ensemble import GradientBoostingClassifier

model2 = GradientBoostingClassifier(
    n_estimators=1,
    max_depth=1,
    min_samples_split=5,
    min_samples_leaf=5,
    learning_rate=0.000001,
)

model2.fit(xtrain, ytrain)
```

# xgboost - Extreme Gradient Boosting.

```python
import xgboost as xgb

xgb_model = xgb.XGBClassifier(max_depth=3)
xgb_model.fit(xtrain, ytrain)
```

```python
scores.mean()
```
