## Step 4 :- Data cleaning
(Handling missing values)
1.categorical(text) - mode (most frequently)
2.continuiuous(numerical)- mean/median

```python
X.columns
```

```python
X.dtypes
```

```python
categorial_cols = list(X.columns[X.dtypes == "object"])
categorial_cols
```

```python
continuous_cols = list(X.columns[X.dtypes != "object"])
continuous_cols
```

```python
df["AirBags"].mode()[0]
```

```python
df["AirBags"].value_counts()
```

```python
df["Luggage.room"].mean()
```

```python
df["Luggage.room"].median()
```

```python
def replacer(df):
    categorial_cols = list(df.columns[df.dtypes == "object"])
    for col in df.columns:

        if col in categorial_cols:
            mode = df[col].mode()[0]
            df[col] = df[col].fillna(mode)
        else:
            median = df[col].median()
            df[col] = df[col].fillna(median)
    print("missing values replaced ")
```

```python
X.isna().sum()
```

```python
replacer(X)
```

### Preprocessing
1.categorial - OneHotEncoding.
2.Continous - StandardScaler.

```python
categorial_cols
```

```python
continuous_cols
```

```python
X_cat = X[categorial_cols]
X_cat
```

```python
X_con = X[continuous_cols]
X_con
```

## Apply one hot encoding on categorial features.

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(
    handle_unknown="ignore", sparse_output=False, drop="first"
).set_output(transform="pandas")
X_cat_pre = ohe.fit_transform(X_cat)
```

```python
X_cat_pre.head()
```

### Continous cols use standardScalar

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler().set_output(transform="pandas")
X_con_pre = scaler.fit_transform(X_con)
```

```python
X_con_pre.head()
```

### Combining X_cat_pre and X_con_pre

```python
X_pre = X_con_pre.join(X_cat_pre)
X_pre.head()
```
